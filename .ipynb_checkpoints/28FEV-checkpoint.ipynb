```python
X = pd.concat([numerical, categorical], axis=1)

y = target['TARGET_B']

```

```python
X
```

```python
from sklearn.model_selection import train_test_split

# Split the data into training and test sets (80% training, 20% testing)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=13)

# Display the shapes of the resulting sets
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

```python
# Identify numerical and categorical columns
numerical_columns = numerical.columns
categorical_columns = categorical.columns

# Split numerical and categorical features for training set
train_num = X_train[numerical_columns]
train_cat = X_train[categorical_columns]

# Split numerical and categorical features for test set
test_num = X_test[numerical_columns]
test_cat = X_test[categorical_columns]

# Display the shapes of the resulting sets
print("train_num shape:", train_num.shape)
print("train_cat shape:", train_cat.shape)
print("test_num shape:", test_num.shape)
print("test_cat shape:", test_cat.shape)
```

```python
for col in categorical.columns:
    print (categorical[col].value_counts(dropna=False))
```

```python
categorical["ODATEW_YR"]=categorical["ODATEW_YR"].replace([85, 84, 83], 86)
```

```python
train_cat=train_cat.drop(["DOB_YR","MINRDATE_YR","MAXRDATE_YR","FIRSTDATE_YR","ODATEW_MM"],axis=1)
train_cat
```

```python
test_cat=test_cat.drop(["DOB_YR","MINRDATE_YR","MAXRDATE_YR","FIRSTDATE_YR","ODATEW_MM"],axis=1)
```

```python
encoder = OneHotEncoder().fit(train_cat)
cols = encoder.get_feature_names_out(input_features=train_cat.columns)

train_cat = pd.DataFrame(encoder.transform(train_cat).toarray(), columns=cols)
test_encoded=encoder.transform(test_cat)

test_cat = pd.DataFrame(encoder.transform(test_cat).toarray(), columns=cols)
```

```python
train_num=train_num.reset_index(drop=True)
test_num=test_num.reset_index(drop=True)
```

```python
# Concatenate scaled numerical and encoded categorical features for both training and test sets
X_train = pd.concat([train_num, train_cat], axis=1)
X_test = pd.concat([test_num, test_cat], axis=1)

# Display the shape of the final X_train and X_test DataFrames
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
```

```python
X_train
```

```python
from sklearn.linear_model import LogisticRegression

classification = LogisticRegression(random_state =0, solver='lbfgs').fit(X_train,y_train)
classification.score(X_test, y_test)
```

```python
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score

predictions = classification.predict(X_test)

print("precision: ",precision_score(y_test,predictions))
print("recall: ",recall_score(y_test,predictions))
print("f1: ",f1_score(y_test,predictions))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,predictions)
```

```python
from sklearn.utils import resample
```

```python
y_train=y_train.reset_index(drop=True)
```

```python
X_train_data = pd.concat([X_train, y_train], axis=1)
X_train_data
```

```python
category_0 = X_train_data[X_train_data['TARGET_B'] == 0]
category_1 = X_train_data[X_train_data['TARGET_B'] == 1]
```

```python
category_1_oversampled = resample(category_1,
                                  replace=True,
                                  n_samples = len(category_0))
```

```python
print(category_0.shape)
print(category_1_oversampled.shape)
```

```python
data_upsampled = pd.concat([category_0, category_1_oversampled], axis=0)
```

```python
data_upsampled['TARGET_B'].value_counts()
```

```python
X_train_upsampled=data_upsampled.drop('TARGET_B',axis=1)
```

```python
y_train_upsampled=data_upsampled['TARGET_B']
```

```python

classification = LogisticRegression(random_state =0, solver='lbfgs').fit(X_train_upsampled,y_train_upsampled)
classification.score(X_test, y_test)
```

```python
predictions = classification.predict(X_test)

print("precision: ",precision_score(y_test,predictions))
print("recall: ",recall_score(y_test,predictions))
print("f1: ",f1_score(y_test,predictions))
```
